```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(1,2,3,4),(5,6,7,8)])

print(pts)

f = plt.figure(figsize=(11.69,8.27))

x_1 = pts[0,:]
y_1 = pts[1,:]

print(x_1,"\n",y_1)

plt.scatter(x_1,y_1)

m, b = np.polyfit(x_1, y_1, 1)

plt.plot(x_1, m*x_1 + b)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(1,2,3,4),(5,6,7,8)])

print(pts)

f = plt.figure(figsize=(11.69,8.27))

x_1 = pts[0,:]
y_1 = pts[1,:]


x_2 = np.delete(x_1,1)
y_2 = np.delete(y_1,2)

print(x_1,"\n",y_1)
print(x_2,"\n",y_2)

plt.scatter(x_1,y_1)
plt.scatter(x_2,y_2)

m1, b1 = np.polyfit(x_1, y_1, 1)
m2, b2 = np.polyfit(x_2, y_2, 1)

plt.plot(x_1, m1*x_1 + b1)
plt.plot(x_2, m2*x_2 + b2)

plt.show()
```
